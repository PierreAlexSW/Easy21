<h2>Easy21's rules</h2>

First let's recall the rules of Easy21's game.
A player (you) is playing against a dealer. An unlimited deck of cards (from 1 to 10 and red or black coloured) is available: red cards account for negative numbers whereas black cards account for positive ones. 

At the beginning of the game, the dealer picks a black card: he starts off with a number between 1 and 10.
Then, it's your turn to pick cards until you lost or decide to stop picking cards (this action is called "stick").
You lose if the sum of your cards is smaller than 1 and greater than 21. 
When you decide to stick, it is dealer's turn to pick. The rules are the same for him.

However, the dealer's policy is known: when his sum is smaller than 17, he continues to pick but if his sum is greater or equal to 17, he sticks (i.e. the game finishes). When the dealer sticks, his sum and yours will be compared: the winner is the one with the largest sum.

<h2>Easy21's MDP</h2>

<h3>Interpretation of the Easy21's MDP as a random walk</h3>

One can interpret the Easy21's game as a random walk.

Given an initial (random) state caracterized by the dealer's card and the player's first card, the player is randomly walking, visiting the possible values taken by the sum of its cards. 
One can better visualize this random walk like the player moving along an axis (horizontal blue axis on the below schema).

When the player's action is "stick", he will stop in that position, then it is dealer's turn. The dealer will then play with his policy, which in our case is known (described in the previous sub-section).
The dealer moves can also be considered as a random walk. He moves along the vertical green axis (see below schema).


WIN/LOSE/DRAW ($\mathcal{W}/\mathcal{L}/\mathcal{D}$) states are terminal states, i.e. the game finishes when landing in such states. If it wasn't clear, WIN is from the point of view of the player.
If the player's sum get out of the $[\![ 1,21]\!]$  interval, he lands on the $\mathcal{L}$ state. During the player's random walk, the $\mathcal{W}$ state is unachievable since a player's win can only happen after the dealer has picked cards (and went out of the interval or sticked). 
Note that the $\mathcal{D}$ state only happens when the dealer has sticked and its sum and the player's are the same.


The MDP is illustrated below:

To sum up, the game can be seen has moving in a matrix of size $(21,21)$ whose rows correspond to the dealer's sum and the columns to the player's sum.
The initial state is a random element from the matrix. When the player "hits", he moves along the same row but when he "sticks", the dealer moves along the same column. When the state of the game is out of this matrix, it means the game is finished and the terminal state is $\mathcal{W}$ or $\mathcal{L}$ or $\mathcal{D}$ (it depends on the last state the game was in).


<h3>The idea behind MDP's solution</h3>

In order to solve Easy21's MDP, one can use <b>value iteration</b>.
Value iteration is an algorithm relying on the following Bellman's equation:
\begin{equation}
    V^{k+1}(s) \leftarrow \max_{a \in A} {\mathcal{R}_{s}^{a} + \mathcal{P}_{s,s'}^{a} V^{k}(s)}
\end{equation}
To use Bellman's equation, we need to compute the expected reward vector and transition matrix for both action "hit" and "stick".

For the sake of clarity, we will now note the considered state $s$, the "hit" action $h$ and the "stick" action $\overline{h}$.

$V(s)$ is the value function at state s, i.e. the expected reward the player can get when he starts at state $s$.
$V(s) = \mathop{\mathbb{E}}\big(\sum_{k=0}^{\infty} \gamma^{k}\mathcal{R}^{k} \, \big| \, s_0 = s\big)$

$\mathcal{R}_{s}^{a}$ is the expected immediate reward when the player takes action $a$ in state $s$, i.e. the reward of this step only. It can take 3 values:
 
$\mathcal{R}_{s}^{a} = 
\begin{cases}
      \text{-}1, & \text{if the player lost}\  \\
      1, & \text{if the player won} \\
      0  &\text{otherwise} 
\end{cases} $

The transition matrix $\mathcal{P}_{s,s'}^{a}$ linked to action $a$ corresponds to the matrix whose elements are the probability to land in state $s'$ when action $a$ was selected from state $s$.

Both the transition matrix and reward linked to the "hit" action can be easily computed since they do not depend on the dealer's initial sum: the player can only loose if he keeps "hitting".

When the action taken is "stick", the rest of the game (dealer's plays) is external to the player since he is doesn't participate to the game any longer. The "stick" action thus blows the game to a terminal state ($\mathcal{W}$ or $\mathcal{L}$ or $\mathcal{D}$): this is the last action of the game.
By definition of the game, this last step is in reality a multiple step random walk where it is the dealer who plays with a certain policy.

$\mathcal{P}_{s,s'}^{\overline{h}}$ is not needed since the expected immediate reward linked to the "stick" action $\mathcal{R}_{s}^{\overline{h}}$ is in fact the overall expected reward (because all previous actions led to 0 reward and the discount factor is one). Moreover, this matrix would be sparse with every non terminal states having null probability to be landed on.

In order to compute $\mathcal{R}_{s}^{\overline{h}}$, <b>one can reduce the two-agent MDP to a one-agent MDP</b>. The initial state of this reduced MDP is the first card of the dealer and the sum of the player when he took action $\overline{h}$ is a parameter of the MDP.

P.S: Make sure to understand the last idea before you continue otherwise you won't understand the following section!!!

<h2>Solving Easy21's MDP</h2>

<h3>Dealer's random walk to get $\mathcal{R}_{s}^{\overline{h}}$</h3>


As suggested in the previous sub-part, when the player has sticked, the dealer's play can be considered as a random walk
whose possible states are the sum of the dealer's cards and $\mathcal{W}/\mathcal{L}/\mathcal{D}$ states (there are 16 + 3 = 19 states).

In this sub-section, we are now considering the <b>one-agent MDP</b> where the dealer is playing alone with his policy. The player's sum is a parameter of the MDP which has an impact on the probabilities to land in terminal states and thus on the overall expected reward.

We will solve this one-agent MDP thanks to <b>value iteration</b> too.
To avoid confusion with the two-agent MDP, we will note $W^k(s)$ the value function at state s at the $k^{th}$ iteration of the algorithm, , $\mathcal{T}_{s}^{a}$ the expected immediate reward and $\mathcal{Q}_{s,s'}^{a}$ the transition matrix of this one-agent MDP. 
As the policy of the dealer is fixed, there is no need to compute the maximum over actions. 

States of this MDP can be written $s = (d, p)$ where $d$ is the sum of the dealer's card (can take values in the $[\![ 1,16]\!]$ interval) and $p$ the sum of the player's card (parameter of the MDP).
We only take into account the states where $d<17$ because we know that landing in a state where the dealer's sum is in $[\![17,21]\!]$ immediately leads to a terminal state.
Since the player's sum $p$ is a parameter of the MDP, the only necessary information caracterizing the state is the dealer's sum $d$. 

The Bellman equation is finally:

\begin{equation}
    W^{k+1}(d) \leftarrow {\mathcal{T}_{d} + \mathcal{Q}_{d,d'} W^{k}(d)}
\end{equation}


Note that in the implementation of the algorithm the $\mathcal{D}$ state is a state that absorbs WIN/LOSE states to avoid looping onto such states and getting infinite rewards.


<h4>$\mathcal{T}_{d}$</h4>


The expected immediate reward vector $\mathcal{T}_{d}$ can be decomposed in two parts: a variable part and a fix part.
The fix part is the part of the expected return linked to the dealer's "hit" action landing him outside the $[\![ 1,21]\!]$ interval.
The variable one describes the part of the expected return where the dealer's sum lands in  $[\![17,21]\!]$: the terminal states depends on the player's sum.


Let's first focus on the fix part which is quite easy since it doesn't depend on the player's sum.
For instance, if the dealer's state is a sum of 1, 10 out of 20 of his possible moves land him on the $\mathcal{D}$ state. Otherwise, he will land on a sum between 2 and 11. 

One can easily find that the fix part $\mathcal{T}_{d}^{fix}$ is the following vector:

\begin{equation}
    \mathcal{T}_{d}^{fix} = \frac{1}{20}
  \begin{pmatrix}
    10 & \cdots & 1 & 0 & 1 & \cdots & 5\\
 \end{pmatrix}^T
\end{equation}

The vector $\mathcal{T}_{d}^{fix}$ above is a vector of size $(16, 1)$.
The fix reward of the $\mathcal{W}/\mathcal{L}/\mathcal{D}$ states (i.e. $ (1, -1,0)^T $) must be concatenated on top of that vector!

Now, let's construct the variable part $\mathcal{T}_{s}^{var}$.

The variable part accounts for the reward the dealer can expect to get when the dealer lands in a state where his sum is between 17 and 21.
Since the dealer can't get a sum $\in [\![17,21]\!]$ if its previous state is between 1 and 6, the variable part for such states is null.

The variable expected return depends on $p$ if $p\in [\![17,21]\!]$.


The table below represents the variable part of the immediate expected return. Be careful, this has to be divided by 20!

| Player/Dealer | 7 | 8 | 9 | 10 | 11 | 12 | $\cdots$ | 16 |
| --- | --- | --- | --- | --- | --- | --- | --- | --- |
| 1 | -1 | -2 | -3 | -4 | -5 | $\cdots$ | $\cdots$ | -5 | 
| $\vdots$ |  $\vdots$| $\vdots$ |  $\vdots$ |  $\vdots$ |  $\vdots$ |  $\vdots$ |  $\vdots$ |  $\vdots$ |
| 16 | -1 | -2 | -3 | -4 | -5 | $\cdots$ | $\cdots$ | -5 | 
| 17 |  0 | -1 | -2 | -3 | -4 | $\cdots$ | $\cdots$ | -4 |
| 18 | 1 | 1 | 0 | -1 | -2 | $\cdots$ | $\cdots$ | -2|
| 19 |  $\vdots$ | 2 | 2 | 1 | 0 | $\cdots$ | $\cdots$ | 0 |
| 20 |  $\vdots$ | $\vdots$ | 3 | 3 | 2 | $\cdots$ | $\cdots$  | 2 |
| 21 | 1 | 2 | 3 | 4 | 4 | $\cdots$ | $\cdots$  | 4 |

To study the MDP with parameter $p$, one just needs to select the $p^{th}$ row of this above table (a $(16,1)$ vector) and to concatenate (on top of that vector) the variable immediate expected reward of the $\mathcal{W}/\mathcal{L}/\mathcal{D}$ states (i.e. $(0, 0,0)^T$). Thus, we get a $(19, 1)$ vector.

$$\mathcal{T}_{d} = \mathcal{T}_{d}^{fix} + \mathcal{T}_{d}^{var}$$
To obtain $\mathcal{T}_{d}$, one just needs to add the fix and variable part. Let's recall that this is a  $(3 + 16, 1)$ vector.



<h4>$\mathcal{Q}_{s,s'}$</h4>

Let's note $\mathcal{Q}_{s,s'}(p)$, the transition matrix of the MDP when the sum of the player is fixed to $p$.

The form of  $\mathcal{Q}_{s,s'}(p)$  is the following:

$\mathcal{Q}_{s,s'}(p) = \begin{pmatrix} {I}_{3\times3} & 0_{3\times16} \\ B_p & A \end{pmatrix}$
where $R_p \in \mathbb{R}^{16\times3}$ and $A \in \mathbb{R}^{16\times16}$
The only part of $\mathcal{Q}_{s,s'}(p)$ that depends on $p$ is $B_p$ since it represents the probability to go from  states $(d,p)$ where $d \in [\![ 1,16]\!] $ to a $(x,p)$ where $x \in \big\{\mathcal{W},\mathcal{L},\mathcal{D}\big\}$ state.
$A$ doesn't depend on $p$ because it corresponds to transition from non-absorbing states to other non-absorbing states.

$A$ is deca-diagonal matrix:
$\begin{equation} A = \frac{1}{20}  \ \begin{pmatrix}0 & 1 & \dotso & 1 &  &  & \\ 1 & \ddots & \ddots & \ddots & \ddots & \huge{0} & \\ \vdots & \ddots & \ddots & \ddots & \ddots & \ddots & \\ 1 & \ddots & \ddots& \ddots & \ddots & \ddots & 1\\& \ddots & \ddots& \ddots & \ddots & \ddots &   \vdots\\ &  \huge{0} & \ddots & \ddots & \ddots & \ddots & 1\\ &  &  & 1 & \dotso & 1 & 0 \\  \end{pmatrix} \end{equation}$

Now, let's find $B_p$!

As we did before to find $\mathcal{T}_{d}$, we seperate $B_p$ in two parts:
$B_p = B_p^{fix} + B_p^{var}$

$B_p^{fix}$ represents the probability that a state $(d,p)$ goes directly into a $\mathcal{W}/\mathcal{L}/\mathcal{D}$ state, i.e. without the player's sum going in the $[\![17,21]\!]$ interval.
In practice, $(d,p)$ can only land in the  $\mathcal{W}$ state (when the dealer's sum is less than 1 or greater than 21).

That's why we have:
$B_p^{fix} = \begin{equation} \frac{1}{20}  \ \begin{pmatrix}10 & 0 & 0   \\ \vdots &\vdots &\vdots \\ 0 & \vdots &\vdots \\ \vdots & \vdots & \vdots \\ 5 & 0 & 0 \end{pmatrix} \end{equation}$


The variable part, $B_p^{variable}$ for all $p$ is the following:


| Dealer / Player(Final state) | 17$(\mathcal{W})$   | 17$(\mathcal{L})$   | 17$(\mathcal{D})$  | 18$(\mathcal{W})$    | 18$(\mathcal{L})$  | 18$(\mathcal{D})$| 19$(\mathcal{W})$    | 19$(\mathcal{L})$  | 19$(\mathcal{D})$| 20$(\mathcal{W})$   | 20$(\mathcal{L})$  | 20$(\mathcal{D})$| 21$(\mathcal{W})$   | 21$(\mathcal{L})$ | 2$(\mathcal{D})$   | Others$(\mathcal{W})$    | Others$(\mathcal{L})$  | Others$(\mathcal{D})$

We can thus construct $V_p$ for each $p$ by adding the fix and variable parts of $V_p$.


<h4>Value iteration</h4>


We can now apply value iteration to solve the MDP (with $p$ a fixed parameter).
The inital value is a zero vector and we apply:

\begin{equation}
    W^{k+1}(d) \leftarrow {\mathcal{T}_{d} + \mathcal{Q}_{d,d'} W^{k}(d)}
\end{equation}

The value iteration algorithms converge to the float precision in less than 200 iterations.
The result of this algorithm is the optimal value function of the MDP with parameter $p$. When one concatenates all $W^{\infty}(d)$ (and remove the first three elements) for all $p$ in $[\![1,21]\!]$, we get a $(21,16)$ matrix.

By reshaping it to a 1D vector of size $(16 \times 21,1)$ then removing all elements for $d>10$ (because these states are not representative in the two-agent MDP) and adding $(1, -1,0)^T$ on top of it, we finally get $\mathcal{R}_{s}^{\overline{h}}$ of the two-agent MDP.

The index of rows of $\mathcal{T}_{d}$ is the following: $\mathcal{W}, \mathcal{L}, \mathcal{D}, (1,1), \cdots, (1,21), (2,1) , \cdots, (2,21), \cdots, (10,21)$.

<h3>Value iteration on the two-agent MDP</h3>

Since we now have $\mathcal{R}_{s}^{\overline{h}}$ (and don't need $\mathcal{P}_{s,s'}^{\overline{h}}$), we just need to construct $\mathcal{R}_{s}^{h}$ and $\mathcal{P}_{s,s'}^{h}$.
Don't worry, it is way easier and faster to do!

<h4>$\mathcal{R}_{s}^{h}$ & $\mathcal{P}_{s,s'}^{h}$</h4>

As the player hits, he may go out of the $[\![ 1,21]\!]$ interval, which means he loses and end up in the $\mathcal{L}$ state.
We have: 


By ordering the matrix so that the index of rows and columns is the following: $\mathcal{W}, \mathcal{L}, \mathcal{D}, (1,1), \cdots, (1,21), (2,1) , \cdots, (2,21), \cdots, (10,21)$, we have:

$\begin{equation}
    \mathcal{R}_{s}^{h} = 
    \begin{pmatrix} C & \cdots & C \\
    \end{pmatrix}^T
\end{equation}$
where $C = - \frac{1}{20} \begin{pmatrix} 10 & \cdots & 0 & \cdots & 10\\
    \end{pmatrix}$



$\begin{equation}
\mathcal{P}_{s,s'}^{h} = \begin{pmatrix}  WLD_1  &  0_{3\times210}\\ WLD_2 & B \\
    \end{pmatrix}
\end{equation}$ where $WLD_1 = \begin{pmatrix}  0  &  0&  1\\ 0  &  0&  1\\0  &  0&  1\\ \end{pmatrix}, WLD_2\in \mathbb{R}^{210\times3}, B \in \mathbb{R}^{210\times210}$

$\begin{equation}
B = \begin{pmatrix} 
    A  &  0_{21\times21}& \cdots & 0_{21\times21}\\ 
    0_{21\times21}& \ddots  &\ddots & \vdots\\
    \vdots & \ddots  & \ddots &  0_{21\times21}\\
    0_{21\times21}& \ddots  &  0_{21\times21} &A\\
    \end{pmatrix}
\end{equation}$
with $A$ is the same deca-diagonal matrix than before:
$\begin{equation} A = \frac{1}{20}  \ \begin{pmatrix}0 & 1 & \dotso & 1 &  &  & \\ 1 & \ddots & \ddots & \ddots & \ddots & \huge{0} & \\ \vdots & \ddots & \ddots & \ddots & \ddots & \ddots & \\ 1 & \ddots & \ddots& \ddots & \ddots & \ddots & 1\\& \ddots & \ddots& \ddots & \ddots & \ddots &   \vdots\\ &  \huge{0} & \ddots & \ddots & \ddots & \ddots & 1\\ &  &  & 1 & \dotso & 1 & 0 \\  \end{pmatrix} \end{equation}$

Finally,
$\begin{equation}
WLD_2 = \begin{pmatrix}   0_{21\times1}  & - C^T & 0_{21\times1} \\ \vdots & \vdots & \vdots \\0_{21\times1}  &  -C^T & 0_{21\times1} \\
    \end{pmatrix}
\end{equation}$ 

<h4>Value iteration</h4>
We finally have everything we need to solve the two-player MDP with value iteration!
The algorithm is the following:
\begin{equation}
    V^{k+1}(s) \leftarrow \max \big\{\mathcal{R}_{s}^{h} + \mathcal{P}_{s,s'}^{h} V^{k}(s), \mathcal{R}_{s}^{\overline{h}}\big\}
\end{equation}

What was added to the value function only considering the "stick" action is a little spike at the states where the player's sum is 11.
One can understand this by the fact that from such states, the probability to land in the $\mathcal{L}$ state is zero.
